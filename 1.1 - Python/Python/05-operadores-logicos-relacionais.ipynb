# Operadores Lógicos

```python
a = True
b = False

c = a and b # a & b
print("'A' e 'B' é iguais a", c)

d = a or b # a | b
print("'A' ou 'B' é iguais a", d)

e = not a 
print("'¬A' é igual a", e)

f = not b 
print("'¬B' é igual a", f)
```
