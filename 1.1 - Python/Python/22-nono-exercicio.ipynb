#Expressões Regulares
*Crie expressões regulares para extrair as seguintes informações do texto abaixo (use a função findall):*
- *Números*
- *CEPs*
- *URLs*

```python
import re
```

```python
texto = 'Minha casa fica na rua Toledo, 76. O CEP é 88388-000 e meu site é https://www.iaexpert.academy'
```

```python
re.findall('\d',texto)
```

```python
re.findall('\d{5}-\d{3}', texto)
```

```python
re.findall('https?://[A-Za-z0-9./]+', texto)
```
