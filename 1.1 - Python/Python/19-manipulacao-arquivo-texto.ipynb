#Manipulação de Arquivos

```python
with open('/content/lorem-ipsum.txt') as texto:
  for linha in texto:
    print(linha) 
```

```python
with open('/content/lorem-ipsum.txt') as texto:
  r = texto.readlines()
```

```python
with open('novo-texto.txt', 'w') as nt:
  nt.write('Lorem ipsum dolor sit amet')
```

```python
with open('novo-texto.txt', 'r') as nt:
  for linha in nt:
    print(linha)
```
