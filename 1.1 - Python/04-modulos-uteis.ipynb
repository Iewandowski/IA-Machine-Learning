# Biblioteca time
*Trabalhar com hora. Ex: medir tempo de execução*

https://docs.python.org/3/library/time.html

```python
import time as tm
```

```python
tm.time()
```

```python
antes = tm.time()
lista = [] 
for i in range(0,10000):
  lista.append(i)
depois = tm.time()

intervalo = depois - antes
print(f'Tempo:: {intervalo} segundos')
```

```python
print('Finalizando.....')
tm.sleep(2)
print('................')
tm.sleep(3)
print('......Finalizado')
```
