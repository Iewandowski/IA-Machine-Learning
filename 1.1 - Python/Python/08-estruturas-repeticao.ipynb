# Estruturas de Repetição - for

```python
for numero in range(1,5):
  print(numero)
```

```python
for numero in range(5,0, -1):
  print(numero)
```

```python
soma = 0
for numero in range(1,6):
  soma = soma + numero
print(soma)
```

```python
palavra = 'palavra'
for letra in palavra:
  print(letra)
```

```python
for i in range(0,5):
  print(i)
  print('----')
  for j in range(0,3):
    print(j)
  print()
```

# Estrutura de Repetição - While

```python
numero = 1
while numero < 6:
  print(numero)
  numero += 1
```

```python
numero = 5
while numero > 0:
  print(numero)
  numero -= 1
```

```python
soma = 0
numero = 1
while numero < 6:
  soma += numero
  numero += 1
print(soma)
```

```python
numero = -1
while numero < 1 or numero > 10:
  numero = int(input('Digite um numero de 1 a 10: '))
```
