#Expressões Regulares:

*   Encontrar as posições de padrões dentro de uma string, se estes estiverem presentes (método ***search***)
*   Encontrar se o começo de uma string é de um determinado padrão (método ***match***)
*   Encontrar todas as substrings em uma string que correspondam a um padrão (método ***findall***)

##Metacaracteres:

*   .   - qualquer caractere (exceto linha nova)
*   \w  - qualquer caractere alfanumérico
*   \W  - qualquer caractere não-alfanumérico
*   \d  - qualquer caractere que seja um dígito (0-9)
*   \D  - qualquer caractere não dígito
*   \s  - espaço em branco
*   ^   - começa com...
*   $   - termina com...
*   "\" - usado antes de metacaracteres para especificar seu significado literal
*   Item da lista

##Quantificadores - 
*Permitem determinar como e quantas vezes os metacaracteres aparecem*


*   []    - opcional entre os que estão dentro dos colchetes
*   ()    - captura grupos de caracteres
*   '*'     - de zero a mais vezes
*   ?     - zero ou uma vez
*   +-    - uma ou mais vezes
*   {m,n} - de m a n vezes
*   |     - ou


##Exemplo:

*   email@hotmail.com
*   email@gmail.com
*   email@yahoo.com.br

Expressão regular que detecta e-mails

```
    \w+@\w+\.w+
```






# Exemplos Práticos:

##Função search

````python
import re
````

````python
frase = 'Olá, meu telefone é (42)0000-0000'
````

````python
re.search('\(\d{2}\)\d{4,5}-\d{4}', frase)
````

````python
frase = 'A placa do carro que eu anotei durante o acidente foi FRT-1998'
````

````python
re.search('[A-Za-z]{3}-\d{4}', frase)
````

````python
email = 'Entre em contato, meu email é teste@teste.com'
````

````python
re.search('\w+@\w+\.com', email)
````

##Função match

````python
frase = 'A placa do carro que eu anotei durante o acidente foi FRT-1998'
````

````python
print(re.match('[A-Za-z]{3}-\d{4}', frase))
````

````python
frase2 = 'FRT-1998 é A placa do carro que eu anotei durante o acidente'
````

````python
print(re.match('[A-Za-z]{3}-\d{4}', frase2))
````

##Função findall

````python
frase3 = 'Meu numero de telefone atual é telefone é (42)0000-0000. O número (56)1111-1111 é o antigo'
````

````python
re.findall('\(\d{2}\)\d{4,5}-\d{4}', frase3)
````
