# Dicionários
*Possui conceito de chave e valor*

```python
coleta = {'Aedes aeypt': 32, 'Aedes albopictus': 22, 'Anopheles darlingi': 14}
```

```python
coleta['Aedes aeypt']
```

```python
coleta['Rhodnius montenegrensis'] = 11
```

```python
print(coleta)
```

```python
del(coleta)['Aedes albopictus']
print(coleta)
```

```python
coleta.items()
```

```python
coleta.keys() # Retorna somente as chaves, sem os valores
```

```python
coleta.values() # Retorna somente os valores, sem as chaves
```

```python
coleta2 = {'Anopheles gambiae': 13, 'Anopheles deaneorum': 14}
print(coleta2)
```

```python
coleta.update(coleta2)
print(coleta)
```

```python
coleta.items()
```

```python
for especies, num_especimes in coleta.items():
  print(f'Espécie: {especies} - Número de espécimes coletados: {num_especimes}')
```

# Conjuntos (set)

```python
biomoleculas = ('proteína', 'ácidos nucleicos', 'carboidrato', 'lipídeo',
                'ácidos nucleicos', 'carboidrato', 'carboidrato', 'carboidrato')
```

```python
print(biomoleculas)
```

```python
print(set(biomoleculas))
```

```python
c1 = {1,2,3,4,5}
c2 = {3,4,5,6,7}
c3 = c1.intersection(c2) # elementos que estão nos dois conjuntos
print(c3)
```

```python
c1.difference(c2) # retorna itens que estão no c1 mas não estão no c2
```
